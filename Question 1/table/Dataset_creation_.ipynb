```python
import os
import requests
from bs4 import BeautifulSoup
import time

# Google image search URL and user-agent
google_image = "https://www.google.com/search?hl=en&tbm=isch&source=hp&biw=1920&bih=1080&"
user_agent = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.77 Safari/537.36"
}

# Folder to save images
saved_folder = 'images_monkey'
if not os.path.exists(saved_folder):
    os.mkdir(saved_folder)

def download_images():
    query = input('What are you looking for? ')
    n_images = int(input('How many images do you want? '))
    print('Searching...')

    images_downloaded = 0
    start = 0
    downloaded_urls = set()  # Set to track downloaded URLs

    while images_downloaded < n_images:
        # Adjust start index for each new request
        search_url = f"{google_image}q={query}&start={start}"
        response = requests.get(search_url, headers=user_agent)

        # Parse HTML
        soup = BeautifulSoup(response.text, 'html.parser')
        image_elements = soup.find_all('img')

        # Collect and download image URLs
        for img in image_elements:
            if images_downloaded >= n_images:
                break
            img_url = img.get('src') or img.get('data-src')
            if img_url and img_url.startswith('http') and img_url not in downloaded_urls:
                try:
                    img_data = requests.get(img_url).content
                    img_name = f"{saved_folder}/{query}_{images_downloaded + 1}.jpg"
                    with open(img_name, 'wb') as img_file:
                        img_file.write(img_data)
                    downloaded_urls.add(img_url)  # Add to set to avoid duplicates
                    images_downloaded += 1
                    print(f"Downloaded image {images_downloaded}")
                    time.sleep(1)
                except Exception as e:
                    print(f"Failed to download image: {e}")

        # Increment start for pagination
        start += 20
        time.sleep(2)  # Delay to prevent rate-limiting

if __name__ == "__main__":
    download_images()
```
